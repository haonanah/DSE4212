```python
import pandas as pd
import certifi
import json
try:
    from urllib.request import urlopen
except ImportError:
    from urllib2 import urlopen

def get_jsonparsed_data(url):
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)

# List of stock tickers
stock_list = ["AAPL", "CVX", "GS", "JNJ", "JPM", "MSFT", "PFE", "PG", "UL", "XOM"]

# API key (replace with your own)
api_key = "SYR3IkXiVZIjJ9WAz69bGcgh4uDMA9m4"

# Loop through each stock and create separate DataFrames for each stock
for stock in stock_list:
    # Create the URL for each stock
    url = f"https://financialmodelingprep.com/api/v3/historical-rating/{stock}?limit=100000&apikey={api_key}"
    
    # Get data for each stock
    data = get_jsonparsed_data(url)
    
    # Convert the data to a pandas DataFrame and dynamically assign it to a variable
    globals()[f"{stock}_df"] = pd.DataFrame(data)




for stock in stock_list:
    # Make sure you're referencing the correct DataFrame variable
    df = globals()[f"{stock}_df"]
    
    # Export the DataFrame to a CSV file
    df.to_csv(f'{stock}_historicalratings.csv', index=False)
    print(f"Exported {stock}_historicalratings.csv")

# Initialize an empty list to store each DataFrame
all_dataframes = []

for stock in stock_list:
    df = globals()[f"{stock}_df"]  # Reference the dynamically created DataFrame
    df['Ticker'] = stock  # Add a column to identify the stock symbol in the concatenated DataFrame
    all_dataframes.append(df)  # Append each DataFrame to the list

# Concatenate all the DataFrames
concatenated_df = pd.concat(all_dataframes, ignore_index=True)

# Optionally, export the concatenated DataFrame to a CSV file
concatenated_df.to_csv('all_stocks_historicalratings.csv', index=False)

# Display the concatenated DataFrame


concatenated_df
```
